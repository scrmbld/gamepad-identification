```python
import matplotlib.pyplot as plt
```

```python
accs = {'90': 0.81, '120': 0.86, '180': 0.88, '240': 0.83}

fig = plt.figure(figsize = (5, 5))

plt.bar(accs.keys(), [x-0.75 for x in list(accs.values())], width=0.5, bottom=0.75)

plt.ylabel('Top 1 Accuracy')
plt.xlabel('Sequence Length')

plt.savefig('results/seq_len_bars')
plt.show()
```
